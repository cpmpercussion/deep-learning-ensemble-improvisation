```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline

```

```python
n64 = pd.DataFrame.from_csv("run_gesture-rnn-4to3-64n3l-201706-loss.csv", index_col=1)
n128 = pd.DataFrame.from_csv("run_gesture-rnn-4to3-128n3l-201706-loss.csv", index_col=1)
n256 = pd.DataFrame.from_csv("run_gesture-rnn-4to3-256n3l-201706-loss.csv", index_col=1)
n512 = pd.DataFrame.from_csv("run_gesture-rnn-4to3-512n3l-201706-loss.csv", index_col=1)

losses = pd.DataFrame(index=n64.index)
# losses.index = n64.index
losses['64'] = n64.Value
losses['128'] = n128.Value
losses['256'] = n256.Value
losses['512'] = n512.Value

total_steps = 1000
steps_per_epoch = total_steps / 50
print(steps_per_epoch)
```

```python
#losses[:].size
plt.style.use('ggplot')
l = losses[:(30*steps_per_epoch)]
```
